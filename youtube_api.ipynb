```python
import requests
import pandas as pd
import time
from secret import api_key
```

```python
def get_video_info(video_id):
    #collection views, comments, and likes
    url_video_stat = 'https://www.googleapis.com/youtube/v3/videos?id='+video_id+'&part=statistics&key='+api_key
    response_video_stats = requests.get(url_video_stat).json()

    view_count = response_video_stats['items'][0]['statistics']['viewCount']
    like_count = response_video_stats['items'][0]['statistics']['likeCount']
    #dislike_count = response_video_stats['items'][0]['statistics']['dislikeCount']
    comment_count = response_video_stats['items'][0]['statistics']['commentCount']
    
    return view_count,like_count,comment_count
```

```python
def get_videos(youtube_df):

    user = 'UCX6OQ3DkcsbYNE6H8uQQuVA' #mrbeast
    #making API call
    pageToken = ''
    url = 'https://www.googleapis.com/youtube/v3/search?key='+api_key+'&channelId='+user+'&part=snippet,id&order=date&maxResults=1000'+pageToken
    #url = 'https://www.googleapis.com/youtube/v3/channels?key='+api+'&forUsername=MrBeast&part=id'

    response = requests.get(url).json()
    time.sleep(1) #to ensure all data has been received before jumping in for loop
    for video in response['items']:
        if video['id']['kind'] == 'youtube#video':
            video_id = video['id']['videoId']
            video_title = video['snippet']['title']
            upload_date = video['snippet']['publishedAt']
            upload_date = upload_date.split('T')[0]

            view_count,like_count,comment_count = get_video_info(video_id)        
            #Using a dictionary to save features to add to DF
            features = {'video_id' : video_id,
                'video_title' : video_title,
                'upload_date' : upload_date,
                'view_count' : view_count,
                'like_count' : like_count,
                'comment_count' : comment_count, 
                }
            
            #concatenate original youtube_df with new DF that contains all features.
            #ignore_index resets index of new DF
            youtube_df = pd.concat([youtube_df, pd.DataFrame(features, index=[0])], ignore_index=True)
    return youtube_df
```

```python
youtube_df = pd.DataFrame(columns=['video_id','video_title', 'upload_date', 'view_count','like_count', 'comment_count'])

youtube_df = get_videos(youtube_df)

```
